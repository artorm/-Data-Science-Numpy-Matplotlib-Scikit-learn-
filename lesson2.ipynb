### Задание 2

```python
t=np.linspace(0, 10, num = 51)
f=np.cos(t)
plt.axis([0.5, 9.5,-2.5, 2.5])
plt.plot(t,f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

###  ЗАдание 3

```python
x=np.linspace(-3, 3, num = 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 =np.sin(x)
fig, ax=plt.subplots(nrows=2, ncols=2)
ax1,ax2,ax3,ax4=ax.flatten()
ax1.plot(x,y1)
ax1.set_title('График y1')
ax1.set_xlim([-5,5])
ax2.plot(x,y2)
ax2.set_title('График y2')

ax3.plot(x,y3)
ax3.set_title('График y3')

ax4.plot(x,y4)
ax4.set_title('График y4')
fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
plt.style.use('fivethirtyeight')
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df['Class'].value_counts()
```

```python
x=df['Class'].value_counts()
x.plot(kind='barh')
plt.xscale(value='log')
plt.show()
```

```python
 dfv1=df.sort_values(['Class'], ascending=False)
```

```python
dfv1.loc[0:492, 'V1']
```

```python
fraud=dfv1.loc[0:492, 'V1']
legally=dfv1.loc[493:, 'V1']

plt.hist(legally, bins=20, density= True, alpha=0.5,
                     histtype='bar', color='gray',
                     edgecolor='white')
plt.hist(fraud, bins=20, density=True, alpha=0.5,
                     histtype='bar', color='red',
                     edgecolor='white')
plt.legend(fontsize = 10,   
           ncol = 2, 
          facecolor = 'oldlace',    
          edgecolor = 'r',   
          title = 'Class 0 - серый/ Class 1 - красный',
                  title_fontsize = '10'   
         )

plt.yscale(value='log')
plt.show()
```
